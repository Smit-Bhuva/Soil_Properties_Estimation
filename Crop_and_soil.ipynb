```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neural_network import MLPRegressor
```

```python
soil_data = pd.read_excel('soil data.xlsx')
soil_data
```

```python
# normalize data

soil_data['Normalized pH'] = (
    (soil_data['pH'] - soil_data['pH'].min())/(soil_data['pH'].max()-soil_data['pH'].min())
)

soil_data['Normalized EC'] = (
    (soil_data['EC dS/m'] - soil_data['EC dS/m'].min())/(soil_data['EC dS/m'].max()-soil_data['EC dS/m'].min())
)

soil_data['Normalized OC'] = (
    (soil_data['OC%'] - soil_data['OC%'].min())/(soil_data['OC%'].max()-soil_data['OC%'].min())
)

soil_data['Normalized P2O5'] = (
    (soil_data['P2O5'] - soil_data['P2O5'].min())/(soil_data['P2O5'].max()-soil_data['P2O5'].min())
)

soil_data['Normalized K'] = (
    (soil_data['K kg/ha'] - soil_data['K kg/ha'].min())/(soil_data['K kg/ha'].max()-soil_data['K kg/ha'].min())
)

soil_data['Normalized S'] = (
    (soil_data['S(ppm)'] - soil_data['S(ppm)'].min())/(soil_data['S(ppm)'].max()-soil_data['S(ppm)'].min())
)

soil_data['Normalized Fe'] = (
    (soil_data['Fe mg/kg'] - soil_data['Fe mg/kg'].min())/(soil_data['Fe mg/kg'].max()-soil_data['Fe mg/kg'].min())
)

soil_data['Normalized Cu'] = (
    (soil_data['Cu mg/kg'] - soil_data['Cu mg/kg'].min())/(soil_data['Cu mg/kg'].max()-soil_data['Cu mg/kg'].min())
)

soil_data['Normalized Mn'] = (
    (soil_data['Mn mg/kg'] - soil_data['Mn mg/kg'].min())/(soil_data['Mn mg/kg'].max()-soil_data['Mn mg/kg'].min())
)

soil_data['Normalized Zn'] = (
    (soil_data['Zn mg/kg'] - soil_data['Zn mg/kg'].min())/(soil_data['Zn mg/kg'].max()-soil_data['Zn mg/kg'].min())
)
```

```python
soil_data
```

```python
# Weights
weights = {
    'pH': 0.15,
    'OC': 0.15,
    'EC': 0.10,
    'P2O5': 0.10,
    'K': 0.15,
    'S': 0.10,
    'Fe': 0.05,
    'Cu': 0.05,
    'Mn': 0.05,
    'Zn': 0.05,
}
```

```python
# Calculate fertility index

soil_data['Fertility index'] = (
    weights['pH'] * soil_data['Normalized pH'] +
    weights['OC'] * soil_data['Normalized OC'] +
    weights['EC'] * soil_data['Normalized EC'] +
    weights['P2O5'] * soil_data['Normalized P2O5'] +
    weights['K'] * soil_data['Normalized K'] +
    weights['S'] * soil_data['Normalized S'] +
    weights['Fe'] * soil_data['Normalized Fe'] +
    weights['Cu'] * soil_data['Normalized Cu'] +
    weights['Mn'] * soil_data['Normalized Mn'] +
    weights['Zn'] * soil_data['Normalized Zn']
)
```

# ML Algorithms 

```python
def load_data(file_path):
    """Load dataset from CSV."""
    return pd.read_excel(file_path)

def preprocess_data(df):
    """Preprocess the data including imputation and scaling."""
    # Impute missing values with median
    imputer = SimpleImputer(strategy='median')
    df_imputed = imputer.fit_transform(df)

    # Scale the data robustly
    scaler = RobustScaler()
    
    df_scaled = scaler.fit_transform(df_imputed)

    return pd.DataFrame(df_scaled, columns=df.columns)

def split_data(df, target_column, test_size=0.3, random_state=42):
    """Split the data into training and testing sets."""
    X = df.drop(columns=[target_column])
    y = df[target_column]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

def train_and_evaluate_models(X_train, X_test, y_train, y_test, column_name):
    """Train and evaluate models for a given column, including both training and testing accuracies."""
    models = {
        "RandomForest": RandomForestRegressor(n_estimators=100, random_state=42),
        "MLPRegressor": MLPRegressor(hidden_layer_sizes=(50, 50), max_iter=10000, random_state=42)
    }

    for name, model in models.items():
        model.fit(X_train, y_train)

        # Calculate and print training accuracy
        train_score = model.score(X_train, y_train)
        print(f"{name} - Training Accuracy: {train_score:.2f}")

        # Predict on the test set
        y_pred = model.predict(X_test)

        # Calculate and print testing accuracy
        test_score = model.score(X_test, y_test)
        print(f"{name} - Testing Accuracy: {test_score:.2f}\n")
```

```python
# Load the dataset
dataframe = load_data('Combined data soil project.xlsx')
dataframe = dataframe.drop('Wavelength', axis=1)

# Preprocess the data
dataframe_preprocessed = preprocess_data(dataframe)
```

```python
# Analyze each column
columns_to_analyze = ['pH', 'EC dS/m', 'OC %', 'Cu mg/kg', 'Mn mg/kg', 'Zn mg/kg', 'Fe mg/kg', 'K kg/ha', 'P2O5', 'S(ppm)']
for column_to_analyze in columns_to_analyze:
    print(column_to_analyze)
    X_train, X_test, y_train, y_test = split_data(dataframe_preprocessed, column_to_analyze)
    train_and_evaluate_models(X_train, X_test, y_train, y_test, column_to_analyze)
```
